```python
import matplotlib.pyplot as plt
import os
```

### SW_EpisodeIV

```python
with open(os.path.join('out_task1/SW_EpisodeIV', 'part-00000')) as f:
    lines = f.readlines()
    characterName = []
    characterCount = []
    for line in lines:
        parts = line.split('\t')
        if len(parts) == 2:
            names = parts[0].strip('\\"')
            count = int(parts[1])
            characterName.append(names)
            characterCount.append(count)


plt.bar(characterName, characterCount)
plt.xlabel('Character Name')
plt.ylabel('Character Count')
plt.title('Top 20 Characters')
plt.xticks(rotation=90)
plt.show()
```
