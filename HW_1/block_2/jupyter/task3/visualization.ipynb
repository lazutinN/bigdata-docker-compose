```python
import matplotlib.pyplot as plt
import os
```

### SW_EpisodeIV

```python
with open(os.path.join('out_task3/SW_EpisodeIV', 'part-00000')) as f:
    lines = f.readlines()
    bigram = []
    bigramCount = []
    for line in lines:
        parts = line.split('\t')
        if len(parts) == 2:
            names = parts[0].strip('\\"')
            count = int(parts[1])
            bigram.append(names)
            bigramCount.append(count)


plt.bar(bigram, bigramCount)
plt.xlabel('Bigram')
plt.ylabel('Bigram Count')
plt.title('Top 20 bigrams')
plt.xticks(rotation=90)
plt.show()
```
